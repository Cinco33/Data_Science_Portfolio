```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
upload = files.upload()
```

```python
#Loading the dataset
df = pd.read_csv('Donors_Data-1.csv')
df.head()
```

```python
# Keep only columns we need
donors = df[['WEALTH', 'HV', 'gender dummy', 'NUMPROM', 'RAMNTALL', 'totalmonths']].copy()

# Rename for simplicity
donors.columns = ['wealth', 'home_value', 'gender', 'promotions', 'total_gifts', 'months_since_last']

# Clean numeric columns (remove $, commas, etc.)
donors['home_value'] = donors['home_value'].replace('[\$,]', '', regex=True).astype(float)
donors['total_gifts'] = donors['total_gifts'].replace('[\$,]', '', regex=True).astype(float)

# Quick check
donors.head()
```

```python
#Q1 Average home value by wealth type

# Drop rows with missing values in 'wealth' or 'home_value'
q1 = donors.dropna(subset=['wealth', 'home_value'])

# Compute averages for home value of each wealth type
q1_results = q1.groupby('wealth', as_index=False)['home_value'].mean()
q1_results = q1_results.rename(columns={'wealth': 'Wealth Type', 'home_value': 'Average Home Value'})

#Display results
q1_results
```

```python
#Gender
q1_gender = q1.groupby(['wealth', 'gender'])['home_value'].mean().round(2).reset_index()

#Compute averages for gender of each wealth type
q1_gender.columns = ['Wealth Type', 'Gender', 'Average Home Value']

#Display results
q1_gender
```

```python
#Q2 Promotions vs total gifts

# Drop rows missing key data
q2 = donors.dropna(subset=['promotions', 'total_gifts'])

# Group and find averages
q2_result = q2.groupby('promotions')[['total_gifts', 'months_since_last']].mean().round(2).reset_index()
q2_result.columns = ['Promotions', 'Average Total Gifts', 'Average Months Since Last Gift']

#Disply results
q2_result
```

```python
#Graphing using Matplot.lib
import matplotlib.pyplot as plt
import numpy as np

#Scatter Plot
plt.scatter(donors['promotions'], donors['total_gifts'], alpha=0.6)
plt.xlabel('Number of Promotions')
plt.ylabel('Total Gifts')
plt.title('Promotions vs Total Giving')
plt.show()
```

I used Python and pandas in Google Colab to analyze the Donors dataset.
After cleaning the data (removing symbols and converting numeric values), I explored two main questions:
Question 1: What is the average home value for each wealth type?
Question 2: How do the number of promotions relate to total giving and recency of donations?
I found that higher wealth types generally have higher average home values.
The second analysis showed that donors who received more promotions tended to give larger total gifts on average.
